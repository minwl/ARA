```python
def Answering(query:str, candidates:list)->str:
  with torch.no_grad():
    answers = extract_reader.predict([query],[candidates])
  return answers['0'][0]["span_answer_text"]
```

```python
app = Flask(__name__)
run_with_ngrok(app)
@app.route("/")
def index():
  return "<h1> Assistant of RA QA </h1>"

@app.route('/QA', methods=['GET'])
def QA():
    if request.method == 'GET':
        query = request.args.get('query')
        candidates = request.args.get('candidates')
        candidates_list = candidates.split("@@@")
        answer = Answering(query, candidates_list)
        return answer

if __name__ == '__main__':
    app.run()
```
